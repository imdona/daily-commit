```python
'''ì„¸ë²ˆì§¸ í’€ì´ - ì—ë¼í† ìŠ¤í…Œë„¤ìŠ¤ì˜ ì²´ í™œìš©í•˜ê¸°
ì‹œê°„ì„ ì¤„ì´ê¸° ìœ„í•´ì„œ ë£¨íŠ¸í•œ ë¶€ë¶„ê¹Œì§€ë§Œ for ë°˜ë³µë¬¸'''
# ê¸°ì¡´ì½”ë“œ
import sys
M, N = map(int, sys.stdin.readline().split())
prime_num = []
for i in range(M, N+1):
    condition = True
    if i == 1 : continue
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            condition = False
            break
    if condition:
        prime_num.append(i)
print(*prime_num, sep='\n')
```
