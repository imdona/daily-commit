```python
class Queue():
    '''í í´ë˜ìŠ¤'''
    def __init__(self):
        '''ìƒì„±ì í•¨ìˆ˜'''
        self.queue = []


    def enqueue(self, item):
        '''íì— ë°ì´í„°ë¥¼ ë„£ëŠ” ë©”ì†Œë“œ'''
        self.queue.append(item)


    def dequeue(self):
        '''íì—ì„œ ë°ì´í„°ë¥¼ êº¼ë‚´ëŠ”(ì‚­ì œ) ë©”ì†Œë“œ : ë§¨ ì²˜ìŒ ì¸ë±ìŠ¤ ê°’'''
        if len(self.queue) == 0:
            return None
        return self.queue.pop(0)


    def return_queue(self):
        '''í ë‚´ë¶€ì— ìˆëŠ” ê°’ì„ ë°˜í™˜í•˜ëŠ” ë©”ì†Œë“œ'''
        return self.queue
```

### 3) ë§í¬ë“œë¦¬ìŠ¤íŠ¸(Linked List) êµ¬í˜„

```python
class Node:
    def __init__(self, value, next = None):
        '''Linked List í´ë˜ìŠ¤ì—ì„œ ì‚¬ìš©í•  Nodeì˜ ìƒì„±ì í•¨ìˆ˜
        value : Nodeì˜ ê°’, next : ìƒì„±ë  Nodeì˜ ë‹¤ìŒ Node, ë””í´íŠ¸ëŠ” None'''
        self.value = value
        self.next = next


class Linkedlist:
    def __init__(self, value):
        '''Linked List ìƒì„±ìí•¨ìˆ˜'''
        self.head = Node(value)


    def add_node(self, value):
        '''Linked Listì— ìƒˆë¡œìš´ Nodeë¥¼ ì¶”ê°€í•˜ëŠ” ë©”ì†Œë“œ'''
        if self.head == None:
            self.head = Node(value)
        else:
            node = self.head
            while node.next:
                node = node.next
            node.next = Node(value)


    def del_node(self, value):
        '''Linked Listì—ì„œ í•´ë‹¹ ê°’(value)ì„ ê°€ì§€ê³  ìˆëŠ” Nodeë¥¼ ì‚­ì œí•˜ëŠ” ë©”ì†Œë“œ'''
        if self.head == None:
            print("í•´ë‹¹ ê°’ì„ ê°€ì§„ ë…¸ë“œê°€ ì—†ì–´ìš”!")
            return  # retrun Noneì´ë‘ ê°™ì€ í‘œí˜„!

        elif self.head.value == value:
            delete = self.head  # delete : ì„ì‹œë…¸ë“œ
            self.head = self.head.next
            del delete
            return value

        else:
            node = self.head
            while node.next:
                if node.next.value == value:
                    delete = self.node.next
                    node.next = node.next.next
                    del delete
                    return value
                else:
                    node = node.next


    def desc(self):
        '''Linked Listì— ì €ì¥ëœ ê°’ë“¤ì„ ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ ë°˜í™˜í•˜ëŠ” ë©”ì†Œë“œ'''
        node = self.head
        result = []

        while node:
            result.append(node.value)
            node = node.next

        return result


    def search_node(self, value):
        '''Linked Listì— í•´ë‹¹ ê°’(value)ì´ ì–´ë””ì— ìœ„ì¹˜í•˜ëŠ”ì§€ ì°¾ëŠ” ë©”ì†Œë“œ'''
        node = self.head
        while node:
            if node.value == value:
                return node

            else:
                node = node.next
```

### 4) ë±(Deque) êµ¬í˜„

```python
class Node:
    def __init__(self, value, next = None):
        '''Deque í´ë˜ìŠ¤ì—ì„œ ì‚¬ìš©í•  Nodeì˜ ìƒì„±ì í•¨ìˆ˜
        value : Nodeì˜ ê°’, next : ìƒì„±ë  Nodeì˜ ë‹¤ìŒ Node, ë””í´íŠ¸ëŠ” None'''
        self.value = value
        self.next = next


class Deque:
    def __init__(self):
        '''ë± í´ë˜ìŠ¤ ìƒì„±ìí•¨ìˆ˜'''
        self.top = None
        self.bottom = None


    def append(self, value):
        '''ë±ì— í•´ë‹¹ ê°’ì„ ì œì¼ ë§ˆì§€ë§‰ ë…¸ë“œì— ì§‘ì–´ë„£ëŠ” ë©”ì†Œë“œ'''
        new_node = Node(value)
        if self.top:
            self.top.next = new_node
            self.top = new_node
        else:
            self.bottom = self.top = new_node


    def appendleft(self, value):
        '''ë±ì— í•´ë‹¹ ê°’ì„ ì œì¼ ì• ë…¸ë“œì— ì§‘ì–´ë„£ëŠ” ë©”ì†Œë“œ'''
        new_node = Node(value)
        if self.bottom:
            new_node.next = self.bottom
            self.bottom = new_node
        else:
            self.bottom = self.top = new_node


    def pop(self):
        '''ë±ì˜ ê°€ì¥ ë’¤ì— ìˆëŠ” ê°’ì„ ì‚­ì œí•˜ëŠ” ë©”ì†Œë“œ'''
        if self.top:
            del_node = self.top
            if self.top == self.bottom:
                self.bottom = self.top = None
            else:
                x = self.bottom
                while x.next:
                    if x.next == self.top:
                        self.top = x
                        break
                    else:
                        x = x.next
            return del_node.value
        else:
            return None


    def popleft(self):
        '''ë±ì˜ ê°€ì¥ ì•ì— ìˆëŠ” ê°’ì„ ì‚­ì œí•˜ëŠ” ë©”ì†Œë“œ'''
        if self.bottom:
            del_node = self.bottom
            self.bottom = del_node.next
            return del_node.value
        else:
            return None


    def ord_desc(self):
        '''ë± ë‚´ë¶€ì— ìˆëŠ” ê°’ì„ ë°˜í™˜í•˜ëŠ” ë©”ì†Œë“œ'''
        out = []
        x = self.bottom
        while x:
            out.append(x.value)
            x = x.next
        return out
```

```python
# ì—°ê²°ë¦¬ìŠ¤íŠ¸ì˜ ë…¸ë“œë¥¼ ì €ì¥í•˜ëŠ” ì²« ë‹¨ê³„

class Node:
    '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ì˜ ë…¸íŠ¸ í´ë˜ìŠ¤'''
    def __init__(self, value, next = None):
        self.value = value  # ë…¸ë“œì˜ ê°’(ë…¸ë“œê°€ ì €ì¥í•˜ëŠ” ë°ì´í„°)ì„ ë‚˜íƒ€ë‚´ëŠ” value
        self.next = next  # ë…¸ë“œì˜ ë‹¤ìŒìœ„ì¹˜ë¥¼ ê°€ë¦¬í‚¤ëŠ” nextì˜ ì´ˆê¸°ê°’ì€ None(ë‹¤ìŒ ë…¸ë“œì— ëŒ€í•œ ë ˆí¼ëŸ°ìŠ¤)

class linked_list:
    def __init__(self, value):
        self.head = Node(value) # ì´ˆê¸° í´ë˜ìŠ¤ì— head ë…¸ë“œ ì„ ì–¸

    def add_node(value):
        print('head.value:', head.value)
        print('head.next:', head.next)
        node = head           # ì²«ë²ˆì§¸ ìœ„ì¹˜ì— í•´ë‹¹í•˜ëŠ” headë¥¼ ìƒì„±í•˜ê³ , node ë³€ìˆ˜ì— ì €ì¥í•´ë‘”ë‹¤.
        while node.next:      # headë…¸ë“œì˜ ë‹¤ìŒìœ„ì¹˜ì— ë…¸ë“œê°€ ìƒì„±ë  ë•Œê¹Œì§€ ë°˜ë³µë¬¸ ì§„í–‰í•œë‹¤.
            node = node.next  # headë…¸ë“œì˜ ë‹¤ìŒìœ„ì¹˜ì— ë…¸ë“œê°€ ìˆëŠ” ê²½ìš°(ë‘ë²ˆì§¸ ë…¸ë“œë¼ê³  ê°€ì •),
            print('print in while:', node.next)
                            # ë‘ë²ˆì§¸ ë…¸ë“œë¶€í„° nodeë³€ìˆ˜ì— ì €ì¥
        node.next = Node(value) # ë°ì´í„°ë¥¼ ë…¸ë“œ ë‹¤ìŒìœ„ì¹˜ì— ì €ì¥
```

##### ğŸ“ íŒŒì´ì¬ ì½”ë“œë¡œ ì—°ê²°ë¦¬ìŠ¤íŠ¸ êµ¬í˜„í•˜ê¸°

```python
# ì—°ê²°ë¦¬ìŠ¤íŠ¸ë„ ë¦¬ìŠ¤íŠ¸ì˜ ë©”ì†Œë“œì™€ ê°™ì€ ê¸°ëŠ¥ë“¤ì„ êµ¬í˜„í•  ìˆ˜ ìˆë‹¤. : ì‚½ì… / ì‚­ì œ / ê²€ìƒ‰
# í´ë˜ìŠ¤ì— ì—°ê²°ë¦¬ìŠ¤íŠ¸ êµ¬í˜„
class Node:
    '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ì˜ ë…¸íŠ¸ í´ë˜ìŠ¤'''
    def __init__(self, value):
        self.value = value
        self.next = None


class linked_list:
    '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ í´ë˜ìŠ¤'''
    def __init__(self, value):
        self.head = Node(value)  # Node í´ë˜ìŠ¤ í˜¸ì¶œ


    def add_node(self, value):
        '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ ì‚½ì… ì—°ì‚° ë©”ì†Œë“œ'''
        if self.head == None:
            # ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ ìˆëŠ” ê²½ìš°
            self.head = Node(value)
        else:
            # ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ ìˆì§€ ì•Šì€ ê²½ìš°
            node = self.head
            while node.next:  # node.nextê°€ Noneì´ ë ë•Œê¹Œì§€ ë°˜ë³µ
                node = node.next
            node.next = Node(value)  # inití•¨ìˆ˜ì˜ value


    # ì—°ê²° ë¦¬ìŠ¤íŠ¸ì˜ ì‚­ì œ êµ¬í˜„
    def del_node(self, value):
        '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ ì‚­ì œ ì—°ì‚° ë©”ì†Œë“œ'''
        # headê°€ Noneì¼ ê²½ìš°
        if self.head == None:
            # í•´ë‹¹ ê°’ì— ëŒ€í•œ ë…¸ë“œëŠ” ì—†ë‹¤
            # ì˜ë¯¸ì—†ëŠ” ì¡°ê±´ì—ì„œ í•¨ìˆ˜ëŠ” ì•„ë¬´ê²ƒë„ ë°˜í™˜í•˜ì§€ ì•ŠëŠ”ë‹¤
            print("í•´ë‹¹ ê°’ì„ ê°€ì§„ ë…¸ë“œê°€ ì—†ì–´ìš”!")
            return

        # headê°€ ì‚­ì œí•´ì•¼í•  ê°’ì¼ ê²½ìš°
        elif self.head.value == value:
            # ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ ë³€ê²½ì‹œí‚¨ë‹¤.
            # ë³€ê²½ëœ ë…¸ë“œì— ëŒ€í•´ì„œ ì‚­ì œë¥¼ ì§„í–‰í•œë‹¤.
            delete = self.head  # ì„ì‹œë…¸ë“œ
            self.head = self.head.next
            del delete

        # ì‚­ì œí•´ì•¼í•  ê°’ ì°¾ê¸°
        else:
            node = self.head  # í˜„ì¬ ìš°ë¦¬ê°€ ë³´ê³  ìˆëŠ” ë…¸ë“œ ì´ˆê¸°í™”í•˜ê¸°

            # ì—°ê²° ë¦¬ìŠ¤íŠ¸ë“¤ì˜ ëª¨ë“  ë…¸ë“œë¥¼ íƒìƒ‰í•˜ê¸°
            while node.next:
                if node.next.value == value:
                    # ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ ë³€ê²½ì‹œí‚¨ë‹¤.
                    # ë³€ê²½ëœ ë…¸ë“œì— ëŒ€í•´ì„œ ì‚­ì œë¥¼ ì§„í–‰í•œë‹¤.
                    delete = node.next
                    node.next = node.next.next  # ë‹¤ìŒì˜ ë‹¤ìŒìœ¼ë¡œ
                    del delete
                    return

                else:
                    # ë‹¤ìŒ ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ í˜„ì¬ ë…¸ë“œì— ë„£ì–´ì¤€ë‹¤. whileë¬¸ì—ì„œ ë°˜ë³µí•˜ê¸° ìœ„í•¨!
                    node = node.next


    # ì—°ê²°ë¦¬ìŠ¤íŠ¸ì˜ ë…¸ë“œì¶œë ¥ì„ ìœ„í•œ ê¸°ëŠ¥
    def ord_desc(self):
        node = self.head
        while node:
            print(node.value)
            node = node.next
```

##### ğŸ“ì—°ê²°ë¦¬ìŠ¤íŠ¸ ê²€ìƒ‰í•¨ìˆ˜ êµ¬í˜„

```python
# ì—°ê²°ë¦¬ìŠ¤íŠ¸ë„ ë¦¬ìŠ¤íŠ¸ì˜ ë©”ì†Œë“œì™€ ê°™ì€ ê¸°ëŠ¥ë“¤ì„ êµ¬í˜„í•  ìˆ˜ ìˆë‹¤. : ì‚½ì… / ì‚­ì œ / ê²€ìƒ‰
# í´ë˜ìŠ¤ì— ì—°ê²°ë¦¬ìŠ¤íŠ¸ êµ¬í˜„
class Node:
    '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ì˜ ë…¸íŠ¸ í´ë˜ìŠ¤'''
    def __init__(self, value):
        self.value = value
        self.next = None


class linked_list:
    '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ í´ë˜ìŠ¤'''
    def __init__(self, value):
        self.head = Node(value)  # Node í´ë˜ìŠ¤ í˜¸ì¶œ


    def add_node(self, value):
        '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ ì‚½ì… ì—°ì‚° ë©”ì†Œë“œ'''
        if self.head == None:
            # ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ ìˆëŠ” ê²½ìš°
            self.head = Node(value)
        else:
            # ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ ìˆì§€ ì•Šì€ ê²½ìš°
            node = self.head
            while node.next:  # node.nextê°€ Noneì´ ë ë•Œê¹Œì§€ ë°˜ë³µ
                node = node.next
            node.next = Node(value)  # inití•¨ìˆ˜ì˜ value


    # ì—°ê²° ë¦¬ìŠ¤íŠ¸ì˜ ì‚­ì œ êµ¬í˜„
    def del_node(self, value):
        '''ë§í¬ë“œ ë¦¬ìŠ¤íŠ¸ ì‚­ì œ ì—°ì‚° ë©”ì†Œë“œ'''
        # headê°€ Noneì¼ ê²½ìš°
        if self.head == None:
            # í•´ë‹¹ ê°’ì— ëŒ€í•œ ë…¸ë“œëŠ” ì—†ë‹¤
            # ì˜ë¯¸ì—†ëŠ” ì¡°ê±´ì—ì„œ í•¨ìˆ˜ëŠ” ì•„ë¬´ê²ƒë„ ë°˜í™˜í•˜ì§€ ì•ŠëŠ”ë‹¤
            print("í•´ë‹¹ ê°’ì„ ê°€ì§„ ë…¸ë“œê°€ ì—†ì–´ìš”!")
            return

        # headê°€ ì‚­ì œí•´ì•¼í•  ê°’ì¼ ê²½ìš°
        elif self.head.value == value:
            # ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ ë³€ê²½ì‹œí‚¨ë‹¤.
            # ë³€ê²½ëœ ë…¸ë“œì— ëŒ€í•´ì„œ ì‚­ì œë¥¼ ì§„í–‰í•œë‹¤.
            delete = self.head  # ì„ì‹œë…¸ë“œ
            self.head = self.head.next
            del delete

        # ì‚­ì œí•´ì•¼í•  ê°’ ì°¾ê¸°
        else:
            node = self.head  # í˜„ì¬ ìš°ë¦¬ê°€ ë³´ê³  ìˆëŠ” ë…¸ë“œ ì´ˆê¸°í™”í•˜ê¸°

            # ì—°ê²° ë¦¬ìŠ¤íŠ¸ë“¤ì˜ ëª¨ë“  ë…¸ë“œë¥¼ íƒìƒ‰í•˜ê¸°
            while node.next:
                if node.next.value == value:
                    # ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ ë³€ê²½ì‹œí‚¨ë‹¤.
                    # ë³€ê²½ëœ ë…¸ë“œì— ëŒ€í•´ì„œ ì‚­ì œë¥¼ ì§„í–‰í•œë‹¤.
                    delete = node.next
                    node.next = node.next.next  # ë‹¤ìŒì˜ ë‹¤ìŒìœ¼ë¡œ
                    del delete
                    return

                else:
                    # ë‹¤ìŒ ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ í˜„ì¬ ë…¸ë“œì— ë„£ì–´ì¤€ë‹¤. whileë¬¸ì—ì„œ ë°˜ë³µí•˜ê¸° ìœ„í•¨!
                    node = node.next


    # ì—°ê²°ë¦¬ìŠ¤íŠ¸ì˜ ë…¸ë“œì¶œë ¥ì„ ìœ„í•œ ê¸°ëŠ¥
    def ord_desc(self):
        node = self.head
        while node:
            print(node.value)
            node = node.next

    # ì—°ê²°ë¦¬ìŠ¤íŠ¸ ê²€ìƒ‰í•¨ìˆ˜
    def search_node(self, value):
        node = self.head
        # ë…¸ë“œê°€ ìˆëŠ” ê²½ìš° ì•„ë˜ ì‘ì—…ì„ ë°˜ë³µí•œë‹¤.
        while node:
            # ë…¸ë“œì˜ ê°’ì´ í˜„ì¬ ê°’ê³¼ ê°™ì€ ê²½ìš°
            if node.value == value:
                return value  #ë…¸ë“œë¥¼ ë°˜í™˜í•œë‹¤.
            # ë…¸ë“œì˜ ê°’ì´ ë‹¤ë¥¸ ê²½ìš°
            else:
                node = node.next  # ë‹¤ë¥¸ ë…¸ë“œì˜ ìœ„ì¹˜ë¥¼ í˜„ì¬ ë…¸ë“œì— ë„£ì–´ì¤€ë‹¤.
```

- ì—°ê²°ë¦¬ìŠ¤íŠ¸ëŠ” íŒŒì´ì¬ì— ì—†ëŠ” ìë£Œêµ¬ì¡°ë¡œ, ì‚¬ìš©í•˜ë ¤ë©´ ì§ì ‘ í´ë˜ìŠ¤ë¡œ êµ¬í˜„í•˜ì—¬ ì‚¬ìš©í•˜ì—¬ì•¼ í•œë‹¤. ì—°ê²°ë¦¬ìŠ¤íŠ¸ë¥¼ êµ¬í˜„í•˜ê³  ì´í•´í•˜ê¸° ìœ„í•´ì„œëŠ” í•´ë‹¹ ìë£Œêµ¬ì¡°ë¥¼ ë§ì´ ì‚¬ìš©í•´ ë³´ì•„ì•¼ í•œë‹¤. ê·¸ë ‡ê¸° ë•Œë¬¸ì— ì—°ê²°ë¦¬ìŠ¤íŠ¸ë¥¼ ê¸°ë°˜ìœ¼ë¡œ íì™€ ìŠ¤íƒ ìë£Œêµ¬ì¡°ë¥¼ êµ¬í˜„í•˜ëŠ” ì—°ìŠµì„ í•´ë´„ìœ¼ë¡œì¨ ì—°ê²°ë¦¬ìŠ¤íŠ¸ë¥¼ ì¢€ ë” ì˜ ì´í•´í•˜ëŠ” ê²ƒì„ ëª©í‘œë¡œ í•œë‹¤.

```python
# ì—°ê²°ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ìš©í•œ í êµ¬í˜„
class LinkedListNode:
    def __init__(self, data):
        self.data = data
        self.next = None


class Queue:
    def __init__(self):
        self.front = None
        self.rear = None


    # ëŒ€ê¸°ì—´ì—ì„œ ë„£ê¸°(íì— ê°’ì„ ì§‘ì–´ë„£ëŠ” í•¨ìˆ˜)
    def enqueue(self, item):
        new_node = LinkedListNode(item)
        # íê°€ ë¹„ì–´ìˆëŠ”ì§€ ì²´í¬
        if self.rear is None:
            self.front = new_node
            self.rear = new_node
        else:
            # ìƒˆë¡œìš´ ë…¸ë“œë¥¼ rear ë‹¤ìŒì— ì‚½ì…
            self.rear.next = new_node
            # ìƒˆë¡œìš´ ë…¸ë“œë¥¼ rear ì¬í• ë‹¹
            self.rear = new_node
        return new_node.data


    # ëŒ€ê¸°ì—´ì—ì„œ ë¹¼ê¸°(íì—ì„œ ê°’ì„ ë¹¼ë‚´ëŠ” í•¨ìˆ˜)
    def dequeue(self):
        # íê°€ ë¹„ì–´ìˆëŠ”ì§€ ì²´í¬
        if self.front is not None:
            # frontê°’ì„ old frontì— ì‚½ì…
            old_front = self.front
            # old front ë‹¤ìŒ ê°’ì„ frontê°’ì— ì‚½ì…
            self.front = old_front.next

        # íê°€ ë¹„ì–´ìˆëŠ”ì§€ ì²´í¬
        if self.front is None:
            # rearë¥¼ Noneìœ¼ë¡œ ì§€ì •í•œë‹¤.
            self.rear = None

        return old_front.data


    # ì—°ê²°ë¦¬ìŠ¤íŠ¸ì˜ ë…¸ë“œì¶œë ¥ì„ ìœ„í•œ ê¸°ëŠ¥
    def ord_desc(self):
        node = self.front
        while node:
            print(node.data)
            node = node.next
```

```python
print('1,2,3 ì°¨ë¡€ëŒ€ë¡œ ì¶”ê°€í•˜ê¸°:')
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.ord_desc()

print('ëŒ€ê¸°ì—´ì—ì„œ ë¹¼ê¸°:')
q.dequeue()
q.ord_desc()
```

```python
# [ì—°ê²°ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ìš©í•œ ìŠ¤íƒ êµ¬í˜„]
# ì•„ë˜ì˜ ì½”ë“œëŠ” ìŠ¤íƒì´ ë‚´ë¶€ì ìœ¼ë¡œ ì–´ë–»ê²Œ ì‘ë™ë˜ëŠ”ì§€ í™•ì¸í•˜ê¸° ìœ„í•´ ì‘ì„±ë˜ì—ˆë‹¤.
# íŒŒì´ì¬ì—ì„œ ìì²´ì ìœ¼ë¡œ ì œê³µí•˜ëŠ” ë¦¬ìŠ¤íŠ¸ì˜ append, pop ë©”ì†Œë“œë¥¼ í™œìš©í•˜ë©´ ì‰½ê²Œ êµ¬í˜„ë  ìˆ˜ ìˆì§€ë§Œ,
# ì•„ë˜ì²˜ëŸ¼ ë‹¤ë¥¸ ë°©ë²•ìœ¼ë¡œë„ êµ¬í˜„í•˜ëŠ” ì½”ë“œë¥¼ ë³´ë©´ì„œ ìŠ¤íƒì— ëŒ€í•œ ì´í•´ë„ë¥¼ ë†’ì¼ ìˆ˜ ìˆë‹¤.

class LinkedListNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        # ì‹ ê·œ ë…¸ë“œ ìƒì„±
        new_node = LinkedListNode(data)
        # ìµœìƒë‹¨ì˜ ë…¸ë“œë¥¼ ì‹ ê·œ ë…¸ë“œ ë‹¤ìŒ ë…¸ë“œë¡œ ì‚½ì…
        new_node.next = self.top
        # ì‹ ê·œ ë…¸ë“œë¥¼ ìµœìƒë‹¨ì— ì‚½ì…
        self.top = new_node
        return new_node.data

    def pop(self):
        # ìŠ¤íƒì´ ë¹„ì–´ìˆëŠ”ì§€ í™•ì¸
        if self.top is not None:
            # ìµœìƒë‹¨ì˜ ë…¸ë“œë¥¼ ì‚­ì œí•  ë…¸ë“œë¡œ ì‚½ì…
            popped_node = self.top
            # ì‚­ì œí•  ë…¸ë“œ ë‹¤ìŒ ë…¸ë“œë¥¼ ìµœìƒë‹¨ì˜ ë…¸ë“œë¡œ ì‚½ì…
            self.top = popped_node.next
            # ì‚­ì œí•  ë…¸ë“œë¡œë¶€í„° ê°’ ë¦¬í„´
            return popped_node.data

    # ì—°ê²°ë¦¬ìŠ¤íŠ¸ì˜ ë…¸ë“œì¶œë ¥ì„ ìœ„í•œ ê¸°ëŠ¥
    def ord_desc(self):
        node = self.top
        while node:
            print(node.data)
            node = node.next
```
